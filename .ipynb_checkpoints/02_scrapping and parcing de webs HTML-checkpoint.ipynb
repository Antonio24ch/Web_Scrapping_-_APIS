## Extracción del código HTML de una web

```python
import urllib.request 

htmlcode = urllib.request.urlopen('https://marca.com/') #código HTML de la pagina marca


for line in htmlcode: 
    print(line.decode().strip())
```

```python
# Importando librerías
import urllib # Para extraer HTML
from bs4 import BeautifulSoup

html = urllib.request.urlopen('https://www.marca.com/')
soup = BeautifulSoup(html) # Extraer codigo transformando al objeto BS

tags  = soup('a') # Extraer todas las etiquetas con hipervinculos y las guarda en tags

# Extraer los valores de atributos HTMl, enlaces a otra web
for tag in tags:
    print('TAG:', tag) # Imprime Etiqueta
    print('URL:', tag.get('href'))  # Devuelve URL
    print('CONTENIDO:', tag.contents) # CONTENTS para imprimir etiqueta
    print('ATRIBUTO:', tag.attrs) # ATTRS imprime los atributos, ejemplo: href = enlaces a otras paginas
# REFERENCIAS DE ATRIBUTOS HTML: https://developer.mozilla.org/es/docs/Web/HTML/Atributos
    print('\n') # Salto de línea 
```
