## Extracción de Texto de una Página Web

```python
import sys
```

```python
# Para trabajar con HTML
import urllib
from bs4 import BeautifulSoup
```

```python
URL = 'https://elpais.com/internacional/2018/01/31/actualidad/1517387619_036241.html'
HTML = urllib.request.urlopen(URL)
BS = BeautifulSoup(HTML)

# Extraer contenido con etiquetas p
tags = BS('p')
```

```python
# Imprimiendo contenido
tags
```

```python
# Filtrando lo que interesa
discurso = ''

# Recorriendo etiwuetas y concatenando los a discurso, siempre que no tenga ningún atributo
for tag in tags:
    if (len(tag.attrs)) ==0: # No quiero que extraiga ese contenido
        a = tag.contents[0] # Extraer contenido, el 0 es para sacar el elemento de la lista
        discurso = discurso + a # El valor a cada una de las etiquetas de tipo p

print(discurso)
```

```python
# Contar palabras para ver cuantas se repiten más

# Crear dict vacío

contador = dict()
# Quitando comas, puntos,etc para contar palabras
discurso = discurso.replace(',', '').replace('.', '').replace('?', '')

# Cortando el discurso por partes
palabras = discurso.split()

for palabra in palabras:
    if len(palabra) >3: # Plaabras mayores a 3
        contador[palabra] = contador.get(palabra,0) + 1 #Get dice cuantas veces aparece la palabra, si no es ninguna se le pone 0 y si ya pareió un 1
        
```

```python
# Imprimiendo diccionario con el contador
contador
```

```python
# Pasarlo a una tabla
import pandas as pd
contador_top_10 = pd.DataFrame(list(contador.items()), columns=['Palabra', 'Contador']).sort_values('Contador', ascending=False)
contador_top_10.head(10)
```
