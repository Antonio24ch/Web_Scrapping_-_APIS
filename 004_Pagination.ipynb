### Extracting Results from Multiple Pages

```python
# Extracting all results from the first pages
# Define an empty list which we will fill with the extracted results
results = []
```

```python
for i in range(5):
    r = requests.get(base_site, params= {'page': i+1})
    
    # Checking if the response is empty 
    if len(r.json()) == 0:
        break
    
    else:
        results.extend(r.json())
```
